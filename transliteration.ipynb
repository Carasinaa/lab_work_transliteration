функция для возвращения пунктуации и регистра к изначальному виду: сранвиваем изначальную строку и транслитерированную строку, если есть различия в рнгистре/пунктуации, убираем их. сохраняем финальный вариант и ретерним его

```python
def origin(text, no_punct_text, translit_text):
  final = ""
  words_ger = text.split()
  words_tr = translit_text.split()
  words_register = no_punct_text.split()
  punctuation = string.punctuation + '»«'
  for i in range(len(words_tr)):
    if words_ger[i][0] in punctuation:
      final += words_ger[i][0]
    if words_register[i][0].isupper():
      final += words_tr[i][0].upper() + words_tr[i][1:]
      if words_ger[i][-1] in punctuation:
        final += words_ger[i][-1] + ' '
      else:
        final += ' '
    else:
      final += words_tr[i]
      if words_ger[i][-1] in punctuation:
        final += words_ger[i][-1] + ' '
      else:
        final += ' '
  return final.strip()
```

функция для транслитерации: убираем регуляркой знаки препинания, понижаем регистр полученной строки. транслитерируем текст, применяем функцию по возвращению регистра и пунктуации в изначальный вид (origin()), ретерним результат

```python
def translit_ger(transitions, text):
    current_state = 'q0'
    translit_text = ""
    new_text = ' ' + re.sub(r'[^\w\s]', ' ', text) + ' '
    for i in new_text.lower():
        if i in transitions[current_state]:
            position = transitions[current_state][i]
            current_state = position[1]
            translit_text += position[0]
    return origin(text, new_text, translit_text)
```
